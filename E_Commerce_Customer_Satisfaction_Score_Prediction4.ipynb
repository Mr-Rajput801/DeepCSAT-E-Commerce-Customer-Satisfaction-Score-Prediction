## Cleaning Data

By looking above analysis we could notice categories such as Unique ID, Customer Remark and Order_ID wont be usefull for further analysis so we suppose to drope those Variables.

Some object variables looks to be like time stamp columns. So the data type need to get change for those variables.

Some Object and float dtype variables has filled with NAN. Those needs to get filled as per needs.


```python
# Drop columns that are not needed for analysis
df1 = df.drop(["Unique id", "Customer Remarks", "Order_id", "order_date_time"], axis=1)

# Display a random sample of 2 rows from the updated DataFrame to verify the result
df1.sample(n=2)
```

```python
# Convert specified columns with object data type to datetime data type
col = ['Issue_reported at', 'issue_responded', 'Survey_response_Date']
for i in col:
    df1[i] = pd.to_datetime(df1[i], dayfirst=True)  # Convert to datetime format with day first
    print(i, ": dtype =", df1[i].dtype)  # Print the column name and its new data type
```

```python
# Fill null values in the DataFrame based on the data type of each column
for i in df1.columns:
    if df1[i].isnull().sum() > 0:  # Check if there are any null values in the column
        if df1[i].dtype == "object":
            # For object (categorical) columns, replace null values with "Unknown"
            df1[i].fillna("Unknown", inplace=True)
            print(i, "Variable has object dtype, so it got Replace its null values to Unknown \n")
        elif df1[i].dtype == "float64":
            # For float columns, replace null values with the median of the column
            df1[i].fillna(df1[i].median(), inplace=True)
            print(i, "Variable has float dtype, so it got Replace its null values to columns median value \n")
```

```python
# Calculate the percentage of missing values and count of unique values for each column in the cleaned DataFrame
null_values1 = round(df1.isnull().sum() * 100 / len(df1), 2)  # Percentage of null values
unique_values1 = df1.nunique()  # Count of unique values in each column

# Create a DataFrame to summarize the analysis of null values, unique values, and data types for each column
analysis_df1 = pd.DataFrame({
    'column_name': df1.columns,         # Column names
    'unique_values': unique_values1,    # Number of unique values
    'null_values_%': null_values1,      # Percentage of null values
    'data_type': df1.dtypes              # Data type of each column
})

# Reset index for the new DataFrame
analysis_df1.reset_index(drop=True, inplace=True)

# Display the summary DataFrame
analysis_df1
```

```python
# Select a subset of columns from the DataFrame `df1`
#retain only the nesessory featurs for the app
df1=df1[["channel_name","Tenure Bucket","Item_price","CSAT Score"]]
```

```python
# Apply one-hot encoding to 'channel_name' and 'Tenure Bucket' (categorical) columns, dropping the first category for each
df1_=pd.get_dummies(df1,columns=["channel_name","Tenure Bucket"],drop_first=True)
```

```python
# Import the StandardScaler from scikit-learn to standardize numerical features
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the 'Item_price' column in the DataFrame to learn its mean and standard deviation
scaler.fit(df1_[["Item_price"]])

# Transform the 'Item_price' column in the DataFrame to have a mean of 0 and a standard deviation of 1
df1_numerical_scaled = scaler.transform(df1_[["Item_price"]])

# Convert the scaled 'Item_price' values back into a DataFrame with the same column name
df1_numerical_scaled = pd.DataFrame(df1_numerical_scaled, columns=["Item_price"])
```

```python
# Combine the one-hot encoded columns with the scaled 'Item_price' and exclude the first column
df1_main = pd.concat([df1_, df1_numerical_scaled], axis=1).iloc[:, 1:]
```

```python
# Import TensorFlow library for building and training machine learning models
import tensorflow

# Import Keras API from TensorFlow for building neural network models
from tensorflow import keras

# Import Sequential model class from Keras for building a linear stack of layers
from tensorflow.keras import Sequential

# Import Dense and Flatten layers for creating and configuring the neural network
from tensorflow.keras.layers import Dense, Flatten
```

```python
# Separate the features (X) from the target variable (y), where 'CSAT Score' is the target
X = df1_main.drop(columns=['CSAT Score'])
y = df1_main['CSAT Score'].values

# Import train_test_split function to divide the dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training (80%) and testing (20%) sets with a fixed random seed for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Import the Sequential model and layer types from Keras for building the neural network
from keras.models import Sequential
from keras.layers import Dense, Flatten, LeakyReLU

# Initialize a Sequential model, which allows stacking layers in a linear fashion
model = Sequential()

# Add a Flatten layer to reshape the input into a 1D array
model.add(Flatten(input_shape=(7, 1)))

# Add the first Dense layer with 128 units followed by a Leaky ReLU activation function
model.add(Dense(128))
model.add(LeakyReLU(alpha=0.01))

# Add additional Dense layers with 128 units and Leaky ReLU activation
model.add(Dense(128))
model.add(LeakyReLU(alpha=0.01))
model.add(Dense(128))
model.add(LeakyReLU(alpha=0.01))
model.add(Dense(128))
model.add(LeakyReLU(alpha=0.01))
model.add(Dense(128))
model.add(LeakyReLU(alpha=0.01))

# Add the output Dense layer with 6 units (one for each class) and a softmax activation function
model.add(Dense(6, activation='softmax'))

# Display the summary of the model architecture, including the number of layers and parameters
model.summary()
```

```python
# Configure the model with a loss function for classification, an optimizer for adjusting learning, and accuracy as a performance metric
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adagrad', metrics=['accuracy'])
```

```python
# Train the model using the training data for 25 epochs and validate its performance on 20% of the training data
history = model.fit(X_train, y_train, epochs=25, validation_split=0.2)
```

```python
# Use the trained model to make predictions on the test data
y_prob = model.predict(X_test)
```

```python
# Convert the predicted probabilities into class labels by selecting the highest probability for each prediction
y_pred = y_prob.argmax(axis=1)
```

```python
# Calculate the accuracy of the model by comparing the predicted labels to the actual labels from the test data
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Import the Gradio library for creating interactive web interfaces
import gradio as gr

# Define a function to make predictions based on user input
def predict(feature1, feature2, feature3, feature4, feature5, feature6, feature7):
    # Convert "True"/"False" inputs to binary values (1 or 0)
    features_np = [
        1 if feature1 == "True" else 0,
        1 if feature2 == "True" else 0,
        1 if feature3 == "True" else 0,
        1 if feature4 == "True" else 0,
        1 if feature5 == "True" else 0,
        1 if feature6 == "True" else 0,
    ]

    # Convert 'Item_price' input to a float and scale it
    feature7 = np.array([[feature7]])
    feature7 = scaler.transform(feature7)[0, 0]

    # Combine features into a single array and reshape for prediction
    features_np = features_np + [feature7]
    features_np = np.array(features_np).reshape(1, -1)

    # Use the model to predict the class based on input features
    prediction = model.predict(features_np)
    predicted_class = np.argmax(prediction, axis=1)
    return str(predicted_class[0])  # Convert the predicted class to a string for display

# Create a Gradio interface to allow users to interact with the prediction function
interface = gr.Interface(
    fn=predict,  # The function that makes predictions
    inputs=[
        gr.Radio(choices=["True", "False"], label="channel_name_Inbound"),  # Radio buttons for 'channel_name_Inbound'
        gr.Radio(choices=["True", "False"], label="channel_name_Outcall"),  # Radio buttons for 'channel_name_Outcall'
        gr.Radio(choices=["True", "False"], label="Tenure Bucket_31-60"),  # Radio buttons for 'Tenure Bucket_31-60'
        gr.Radio(choices=["True", "False"], label="Tenure Bucket_61-90"),  # Radio buttons for 'Tenure Bucket_61-90'
        gr.Radio(choices=["True", "False"], label="Tenure Bucket_>90"),  # Radio buttons for 'Tenure Bucket_>90'
        gr.Radio(choices=["True", "False"], label="Tenure Bucket_On Job Training"),  # Radio buttons for 'Tenure Bucket_On Job Training'
        gr.Textbox(label="Item_price", type="text")  # Textbox for entering 'Item_price'
    ],
    outputs=gr.Textbox(label="CSAT Score"),  # Display the predicted CSAT score
    title="eCommerce Customer Service Satisfaction App",  # Title of the app
    description="Please enter all the values to predict CSAT Score"  # Description of the app
)

# Launch the Gradio interface so users can interact with the app
interface.launch()

```
